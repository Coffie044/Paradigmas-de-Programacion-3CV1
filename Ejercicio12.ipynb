```python
class SaldoEfectivoInsuficiente(Exception):
    pass
```

```python
class SaldoCuentaInsuficiente(Exception):
    pass
```

```python
class CuentaHabiente:
    def __init__(self, nombre_usuario, contraseña, saldo):
        self.nombre_usuario=nombre_usuario
        self.contraseña=contraseña
        self.saldo=saldo
```

```python
class CajeroAutomatico:
    def __init__(self):
        self.efectivo_cajero=100000
        self.cuentas={}
    def agregar_cuenta(self, cuentahabiente):
        self.cuentas[cuentahabiente.nombre_usuario]=cuentahabiente
    def autenticar(self, nombre_usuario, contraseña):
        cuentahabiente=self.cuentas.get(nombre_usuario)
        if(cuentahabiente and cuentahabiente.contraseña==contraseña):
            return cuentahabiente
        return None
    def mostrar_saldo(self, cuenta):
        return cuenta.saldo
    def depositar_a_propia(self, cuenta, monto):
        cuenta.saldo+=monto
        self.efectivo_cajero+=monto
        return f"Depósito exitoso. Nuevo saldo: {cuenta.saldo}"
    def depositar_a_otra(self, cuenta_destino_num, monto):
        cuenta_destino = self.cuentas.get(cuenta_destino_num)
        if(cuenta_destino):
            cuenta_destino.saldo+=monto
            self.efectivo_cajero+=monto
            return f"Depósito exitoso a la cuenta de {cuenta_destino_num}. Nuevo saldo: {cuenta_destino.saldo}"
        else:
            return "Cuenta destino no encontrada"
    def transferir_a_otra(self, cuenta_origen, nombre_usuario_destino, monto):
        cuenta_destino = self.cuentas.get(nombre_usuario_destino)
        if cuenta_destino:
            if(cuenta_origen.saldo>=monto):
                cuenta_origen.saldo-=monto
                cuenta_destino.saldo+=monto
                return f"Transferencia exitosa a la cuenta de {nombre_usuario_destino}. Nuevo saldo en cuenta origen: {cuenta_origen.saldo}"
            else:
                raise SaldoCuentaInsuficiente("Saldo insuficiente en la cuenta de origen")
        else:
            return "Cuenta destino no encontrada"
    def retirar_efectivo(self, cuenta, monto):
        if(cuenta.saldo>=monto):
            if(self.efectivo_cajero>=monto):
                cuenta.saldo-=monto
                self.efectivo_cajero-=monto
                return f"Retiro exitoso. Nuevo saldo: {cuenta.saldo}, Efectivo en cajero: {self.efectivo_cajero}"
            else:
                raise SaldoEfectivoInsuficiente("Efectivo insuficiente en el cajero")
        else:
            raise SaldoCuentaInsuficiente("Saldo insuficiente en la cuenta")
```

```python
def menu():
    cajero=CajeroAutomatico()
    cuenta1=CuentaHabiente("usuario1", "1234", 5000)
    cuenta2=CuentaHabiente("usuario2", "5678", 3000)
    cajero.agregar_cuenta(cuenta1)
    cajero.agregar_cuenta(cuenta2)
    print("Cajero Automatico")
    nombre_usuario=input("Ingrese su nombre de usuario: ")
    contraseña=input("Ingrese su contraseña: ")
    cuenta=cajero.autenticar(nombre_usuario, contraseña)
    if(cuenta):
        print(f"Bienvenido {cuenta.nombre_usuario}")
        while(True):
            print("\nSeleccione una operación:")
            print("1. Mostrar saldo")
            print("2. Depositar en cuenta propia")
            print("3. Depositar en otra cuenta")
            print("4. Transferir a otra cuenta")
            print("5. Retirar efectivo")
            print("6. Salir")
            opcion=input("Ingrese el número de la operación: ")            
            if(opcion=="1"):
                print(f"Saldo actual: {cajero.mostrar_saldo(cuenta)}")
            elif(opcion=="2"):
                monto=float(input("Ingrese el monto a depositar: "))
                print(cajero.depositar_a_propia(cuenta, monto))
            elif(opcion=="3"):
                nombre_usuario_destino=input("Ingrese el nombre de usuario destino: ")
                monto=float(input("Ingrese el monto a depositar: "))
                print(cajero.depositar_a_otra(nombre_usuario_destino, monto))
            elif(opcion=="4"):
                nombre_usuario_destino=input("Ingrese el nombre de usuario destino: ")
                monto=float(input("Ingrese el monto a transferir: "))
                try:
                    print(cajero.transferir_a_otra(cuenta, nombre_usuario_destino, monto))
                except SaldoCuentaInsuficiente as e:
                    print(e)
            elif(opcion=="5"):
                monto=float(input("Ingrese el monto a retirar: "))
                try:
                    print(cajero.retirar_efectivo(cuenta, monto))
                except (SaldoEfectivoInsuficiente, SaldoCuentaInsuficiente) as e:
                    print(e)
            elif(opcion=="6"):
                print("Hasta pronto.")
                break
            else:
                print("Opción no válida. Intente de nuevo.")
    else:
        print("Autenticación fallida. Verifique su nombre de usuario y contraseña.")
```

```python
if __name__=="__main__":
    menu()
```
